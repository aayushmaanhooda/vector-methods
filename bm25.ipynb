```python
'''
PROBLEM IN TF-IDF

Doc A: "Python Python Python ... Python" (word repeated 100 times)
Doc B: "Python is great for beginners" (word once)

TF-IDF says Doc A is 10× better. But is it really?
Maybe Doc B is actually better content!
'''
```

```python
'''
SOLUTION


1. SATURATION - After a word appears enough times, extra repetitions don't help much.
                
                Appearances: 1 → 2 → 5 → 10 → 100
                Score boost:  +50% → +30% → +10% → +2%

                After 10 times, 100 times doesn't help much more

                Doc A: "Python Python Python Python Python" (5 times)
                Score = 2.5

                Doc B: "Python Python Python Python Python Python Python Python Python Python" (10 times)
                Score = 2.8

                Only +0.3 boost, not 2× boost!


2. Document length normalization - Long documents naturally have more words. BM25 adjusts for this.
                                    Doc A: "Python" (10 words total)
                                    TF = 1/10 = 0.1

                                    Doc B: "Python is a great language for data science..." (1000 words)
                                    TF = 1/1000 = 0.001

                                    BM25 says: Wait, Doc A is proportionally more about Python!
                                    Adjusts scoring so Doc A ranks higheR


3. Parameter tuning
BM25 has adjustable knobs:

k1: Controls saturation (default: 1.2)
b: Controls document length impact (default: 0.75)

You can tune these for your data.
'''
```
