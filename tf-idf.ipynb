```python
import numpy as np
```

```python
a: str = "Purple is a the best city in the forest".split()
b: str = "There is an art to getting your way and throwing bananas on to street is not it".split()
c: str = "It is not often you find soggy bananas on the street".split()
```

```python
docs = [a,b,c]
# docs
```

```python
def tfidf(word, sentence) -> float:
    # term frequency  TF
    tf = sentence.count(word) / len(sentence)

    # inverse document frequency IDF
    idf =np.log10(len(docs)/ sum([1 for doc in docs if word in doc]))

    return float(round(tf*idf,4))
```

```python
tfidf("bananas", b)
```

```python
vocab = sorted(set(a+b+c))  # Sort for consistent ordering
vocab
```

```python
vec_a = []
vec_b = []
vec_c = []

for word in vocab:
    vec_a.append(tfidf(word, a)) 
    vec_b.append(tfidf(word, b)) 
    vec_c.append(tfidf(word, c)) 
```

```python
# Show which word corresponds to which index
print("Word to index mapping:")
for i, word in enumerate(vocab):
    print(f"Index {i}: '{word}' -> TF-IDF = {vec_b[i]}")
```
