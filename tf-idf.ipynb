```python
import numpy as np
```

```python
a: str = "Purple is a the best city in the forest".split()
b: str = "There is an art to getting your way and throwing bananas on to street is not it".split()
c: str = "It is not often you find soggy bananas on the street".split()
```

```python
docs = [a,b,c]
# docs
```

```python
def tfidf(word, sentence) -> float:
    # term frequency  TF
    tf = sentence.count(word) / len(sentence)

    # inverse document frequency IDF
    idf =np.log10(len(docs)/ sum([1 for doc in docs if word in doc]))

    return round(tf*idf,4)
```

```python
tfidf("bananas", b)
```
